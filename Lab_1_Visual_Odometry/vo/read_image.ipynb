```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
def read_depth(filename):
    with open(filename, "r") as f:     
        data = (f.read())
        data = data.split(' ')[:-1]
        data = [float(x)  for x in data]
        depth0 = np.reshape(data,(480,640))
        return depth0
```

```python
frame0 = cv2.imread('image/img1250_0.png')
```

```python
depth0 = read_depth('depth/img1250_0.depth')
```
